Clean the files to remove the disclaimer

```python
def clean_book(document):
    # split the document into lines
    lines = document.split("\n")
    start = 0
    end = len(lines)
    # identify the start and the end of the disclaimer
    for i in range(len(lines)):
        line = lines[i]
        if line.startswith("*** START OF THIS PROJECT GUTENBERG"):
            start = i+1
        elif line.startswith("*** END OF THIS PROJECT GUTENBERG"):
            end = i-1
    # recreate the book without the disclaimer
    return "\n".join(lines[start:end])
```

Create a function that loads all of the books, performs the preprocessing, and returns them along with a class number for each author

```python
import numpy as np
def load_books_data(folder=data_folder):
    # lists for storing the documents themselves and the author classes
    documents = []
    authors = []
    subfolders = [subfolder for subfolder in os.listdir(folder) if 
                 os.path.isdir(os.path.join(folder, subfolder))]
    for author_number, subfolder in enumerate(subfolders):
        # create the full subfolder path and look for all documents within
        # that subfolder
        full_subfolder_path = os.path.join(folder, subfolder)
        for document_name in os.listdir(full_subfolder_path):
            with open(os.path.join(full_subfolder_path, document_name),'rb') as inf:
                inf = inf.read().decode(errors='replace')
                documents.append(clean_book(inf))
                authors.append(author_number)
    # return the documents and the classes (transformed into Numpy array)
    return documents, np.array(authors, dtype='int')
documents, classes = load_books_data(data_folder)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
extractor = CountVectorizer(vocabulary=function_words)
from sklearn.svm import SVC
from sklearn.cross_validation import cross_val_score
from sklearn.pipeline import Pipeline
from sklearn import grid_search
parameters = {'kernel':('linear', 'rbf'), 'C':[1, 10]}
svr = SVC()
grid = grid_search.GridSearchCV(svr, parameters)

pipeline1 = Pipeline([('feature_extraction', extractor),
('clf', grid)
])

#pipeline1

#scores = cross_val_score(pipeline1, documents, classes)
#print("Score: {:.3f}".format(np.mean(scores)))

scores = cross_val_score(pipeline1, documents, classes, scoring='accuracy')
print("Score: {:.3f}".format(np.mean(scores)))
```

n-gram counts

```python
pipeline = Pipeline([('feature_extraction', CountVectorizer(analyzer='char',
ngram_range=(3, 3))),('classifier', grid)
])
                      
scores = cross_val_score(pipeline, documents, classes, scoring='precision_micro')
print("Score: {:.3f}".format(np.mean(scores)))
```

```python
from sklearn.cross_validation import train_test_split
training_documents, testing_documents, y_train, y_test = train_test_split(documents, classes, random_state=14)
pipeline.fit(training_documents, y_train)
y_pred = pipeline.predict(testing_documents)
print(pipeline.named_steps['classifier'].best_params_)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_pred, y_test)
cm = cm / cm.astype(np.float).sum(axis=1)
```
